```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline

print("Imported!")
```

```python
pact_ww = pd.read_csv("pact+pinfo.csv", sep="|")
print pact_ww.columns
sess_ww = pd.read_csv("sess+pinfo.csv", sep="|")
print sess_ww.columns
pinfo_ww = pd.read_csv("pinfo_ww.csv", sep="|")
print pinfo_ww.columns
print "Readed!"
```

```python
#количество игроков;
table_count_players = pinfo_ww.groupby('city').size()
table_count_players.sort_values(inplace=True)
print table_count_players.tail(20)
table_count_players.sort_values().tail(20).plot(kind='bar')
```

```python
#количество игроков в среднем;
table_count_players_mean = pinfo_ww.groupby('city').size().mean()
print table_count_players_mean
```

```python
#количество платежей (transactions);
table_transactions = pact_ww.groupby(['city']).transactions.sum()
print table_transactions.sort_values(inplace=False).tail(20)
table_transactions.sort_values(inplace=False).tail(20).plot(kind='bar')
```

```python
#количество платежей (transactions) (в среднем);
table_transactions = pact_ww.groupby(['city']).transactions.mean()
print table_transactions.sort_values(inplace=False).tail(20)
table_transactions.sort_values(inplace=False).tail(20).plot(kind='bar')
```

```python
#количество пройденных квестов;
table_quests_end = pact_ww.groupby('city').quests_end.sum()
table_quests_end.sort_values(inplace='True')
print table_quests_end.tail(20)
table_quests_end.tail(20).plot(kind='bar')
```

```python
#количество пройденных квестов (в среднем);
table_quests_end = pact_ww.groupby('city').quests_end.mean()
table_quests_end.sort_values(inplace='True')
print table_quests_end.tail(20)
table_quests_end.tail(20).plot(kind='bar')
```

```python
#длина сессии (средняя);
table_session_length = sess_ww.groupby(['city']).session_length.mean()
table_session_length.sort_values(inplace=True)
print table_session_length.tail(50)
table_session_length.tail(20).plot(kind='bar')
```

```python
#количество пройденных сайд-кластеров;
#ml_num - Номер кластера основной сюжетной линии
ml_max_num = pact_ww.ml_num
ml_max_num = ml_max_num.sort_values()
#print ml_max_num.max() #максимальный кластер

table_ml_num = pact_ww.groupby(['id','city'], as_index=False).ml_num.max()
table_ml_num = table_ml_num.groupby('city', as_index=True).ml_num.mean()
table_ml_num = table_ml_num.sort_values()
print table_ml_num.tail(20)
table_ml_num.tail(20).plot(kind = "bar")
```
